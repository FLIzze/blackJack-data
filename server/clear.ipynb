```python
import pandas as pd

df= pd.read_csv('filterProba.csv')

df
```

```python
df['WinPercentage'] = round((df['Win'] / df['Total']) * 100,2)

df['GroupID'] = df.index // 3
df

winning_rows = df.loc[df.groupby('GroupID')['WinPercentage'].idxmax()]

winning_rows

winning_rows.to_csv('winning_rows.csv', index=False)
```
